### Linear Regression on entire dataset

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(df[['R&D Spend','Administration','Marketing Spend']],y=df['Profit'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[10000,20000,40000]])
```

```python
model.score(df[['R&D Spend','Administration','Marketing Spend']],y=df['Profit'])
```

**Conclusion-Accuracy for entire dataset is 95.07%**

### Splitting into train_test data

```python
x=df[['R&D Spend','Administration','Marketing Spend']]
x
```

```python
y=df[['Profit']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
x_pred=model.predict(x_train)
x_pred
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
